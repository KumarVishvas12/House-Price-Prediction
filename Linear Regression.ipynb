```python
n=len(Y)
def initliz_param(Y):
    m=0.1
    c=Y.mean()
    return m,c
```

```python
def generate_pred(m,c,X):
    prediction=[]
    for x in X:
        prediction.append(x*m+c)
    return prediction
```

```python
def calc_cost(prediction,Y,n):
    return (np.sum((prediction-Y)**2))/n
```

```python
def gradient(prediction,Y,X,n):
    Gm=2/n*np.sum((prediction-Y)*X)
    Gc=2/n*np.sum((prediction-Y))
    return Gm,Gc
```

```python
def update_param(m_old,c_old,Gm,Gc,alpha):
    m_new=m_old-alpha*Gm
    c_new=c_old-alpha*Gc
    return m_new,c_new
```

```python
alpha=0.01
max_iter=1000
m,c=initliz_param(sale_price)
cost_old=0
cost=[]
for i in range(max_iter):
    prediction=generate_pred(m,c,plot_area)
    
    cost_new=calc_cost(prediction,sale_price,n)
    if abs(cost_new-cost_old)<10**(-7):
        break
    Gm,Gc=gradient(prediction,sale_price,plot_area,n)
    m,c=update_param(m,c,Gm,Gc,alpha)
    cost.append(cost_new)
    cost_old=cost_new
k=range(max_iter)  
plt.figure(figsize=(16,9))
plt.scatter(plot_area,sale_price,color='red')  
plt.plot(plot_area,prediction,color='green')
```

```python
VIF[VIF==VIF.max()].keys()
```

```python
def remove_multicollinearity_by_vif(data):
    
    VIF=pd.Series([variance_inflation_factor(vif_data.values,i) for i in range(vif_data.shape[1])],index=vif_data.columns)
    if VIF.max()>5:
        print(VIF[VIF==VIF.max()].keys()," is droped")
        vif_data.drop(columns=VIF[VIF==VIF.max()].keys(),inplace=True)
        return vif_data
        
    else:
        print("Vif for column is not greater than 5")
        return vif_data
for i in range(5):
    vif_data=remove_multicollinearity_by_vif(vif_data)
```
